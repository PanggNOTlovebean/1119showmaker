```python
import pandas as pd
df1=pd.read_csv('20日相关.csv')
df2=pd.read_csv('20日涨幅.csv')
df3=pd.read_csv('配对交易.csv')
```

```python
import numpy as np
avg_list=[]
for column in df1.columns:
    avg=np.mean(df1[column]['2018-12-03':])
    avg_list.append(avg)
len(avg_list)
```

```python
profit_list=[]
for column in df3.columns:
    profit=df3[column].iloc[-1]
    profit_list.append(profit)
len(profit_list)
```

```python
x=np.array(avg_list)
y=np.array(profit_list)
y=(y-20000)/20000
print(x.min(),x.max())
np.linspace(start=0.4,stop=0.8,num=5)
r1=[]
r2=[]
r3=[]
r4=[]
r5=[]
for i in range(len(x)):
    if(y[i]!=0):
        if(x[i]<0.4):
            r1.append(y[i])
        elif(x[i]<0.5):
            r2.append(y[i])
        elif(x[i]<0.6):
            r3.append(y[i])
        elif(x[i]<0.7):
            r4.append(y[i])
        elif(x[i]<0.8):
            r5.append(y[i])
y=[np.mean(r1),np.mean(r2),np.mean(r3),np.mean(r4),np.mean(r5)]

# 28+365+365-30-31=697
y=np.array(y)/697*365
print(y)
```

```python
import matplotlib.pyplot as plt
plt.rcParams["font.family"] = ["sans-serif"]
plt.rcParams["font.sans-serif"] = ["SimHei"]
plt.rcParams["axes.unicode_minus"] = False
plt.figure()
xname=['0.3~0.4','0.4~0.5','0.5~0.6','0.6~0.7','0.7~0.8']
plt.plot(xname,y,marker='*',ms=12)
plt.xlabel('相关性均值')
plt.ylabel('收益率')
plt.show()
```
